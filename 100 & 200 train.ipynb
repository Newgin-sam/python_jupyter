```python
import numpy as np
import pandas as pd
from statistics import mean
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcdefaults()
```

```python
def box_plot_comparison(data,title='',xlabel='',ylabel=''):
    # Set the figure size
    plt.rcParams["figure.figsize"] = [12.00, 5.00]
    plt.rcParams["figure.autolayout"] = True

    df = pd.DataFrame(data)

    # Initialize lists to store the spread data
    types = []
    methods = []
    values = []

    # Loop through the data and extract values
    for item in data:
        for val in item["data"]:
            types.append(item["type"])
            methods.append(item["method"])
            values.append(val)

    # Create a DataFrame from the spread data
    df = pd.DataFrame({"type": types, "method": methods, "score": values})

    mean_df = df.groupby(['type', 'method'])['score'].mean().reset_index()

    print(df)

    # Create a figure with two y-axes
    fig, ax = plt.subplots()

    # Create a grouped boxplot
    sns.boxplot(
        x=df['type'],
        y=df['score'],
        hue=df['method'],
        showmeans=True,
        meanprops={
            'marker': 'o',
            'markerfacecolor': 'white',
            'markeredgecolor': 'black',
            'markersize': 3
        },
        ax=ax  # Specify the first y-axis
    )

    ax.set(
        xlabel=xlabel,
        ylabel =ylabel
    )

    ax.set_title(title)
    plt.legend(title='')

#     plt.show()
    return plt
```
